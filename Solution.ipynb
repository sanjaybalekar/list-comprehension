# Find all of the numbers from 1–1000 that have a 6 in them

```python
l=[i for i in range(1,1000) if str(i).find('6')>=0]
print(l)
```

# Find all of the numbers from 1–1000 that are divisible by 8

```python
l=[i for i in range(1,1000) if i%8==0]
print(l)
```

# Count the number of spaces in a string

```python
s = 'Practice Problems to Drill List Comprehension in Your Head.'
l = [i for i in s if i==' ']
len(l)
```

# Remove all of the vowels in a string

```python
s = 'Practice Problems to Drill List Comprehension in Your Head.'
l = [i for i in s if i not in ('a','e','i','o','u')]
''.join(l)
```

# Find all of the words in a string that are less than 5 letters

```python
s = 'Practice Problems to Drill List Comprehension in Your Head.'
l = [i for i in s.split() if len(i)<5]
l
```

# Use a nested list comprehension to find all of the numbers from 1–1000 that are divisible by any single digit besides 1 (2–9)

```python
l = [ i for i in range(1,1000) for j in range(2,10) if i%j==0]
print(set(l))
```
